```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

np.warnings.filterwarnings('ignore')
```

```python
test_dataset=pd.read_csv("./Data/Test.csv")
a=test_dataset['Item_Identifier']
test_dataset['Source']=['Test' for x in a]
```

```python
train_dataset=pd.read_csv("./Data/Train.csv")
a=train_dataset['Item_Identifier']
train_dataset['Source']=['Train' for x in a]
```

```python
frame=[test_dataset,train_dataset]
concat_data=pd.concat(frame,sort=True,ignore_index=True)
concat_data
```

```python

concat_data.describe()
```

```python
for column in list(concat_data):
    
    count=sum(concat_data[column].isnull().values)
    print("Number of missing values in %s is %s"%(column,count))
    print()
```

```python
for column in list(concat_data):
    k=len(concat_data.groupby(column)[column].nunique())
    print("unique value of %s is %s"%(column,k))
    
```

```python
for column in list(concat_data):
    p=(concat_data.groupby(column)[column])
    
    if(column=="Item_Fat_Content" or column=="Item_Type" or column=="Outlet_Size" or column=="Outlet_Location_Type" or column=="Outlet_Type"):
        print(column)
        for city,city_df in p:
            print(city," = ",len(city_df))
        print()
```

```python
print(sum(concat_data['Item_Weight'].isnull().values))
mean_weight = concat_data.pivot_table(values="Item_Weight",index="Item_Identifier")
miss_val = concat_data['Item_Weight'].isnull()
concat_data.loc[miss_val,'Item_Weight'] = concat_data.loc[miss_val,'Item_Identifier'].apply(lambda x:mean_weight.loc[x])
print(sum(concat_data['Item_Weight'].isnull().values))
```

```python
from scipy.stats import mode
t=concat_data.pivot_table(values='Outlet_Size',columns = 'Outlet_Type',aggfunc=(lambda x:mode(x.astype('str')).mode[0]))
t['Grocery Store']['Outlet_Size']='Small'
miss_val = concat_data['Outlet_Size'].isnull()
print(sum(miss_val))

concat_data.loc[miss_val,'Outlet_Size'] = concat_data.loc[miss_val,'Outlet_Type'].apply(lambda x: t[x])

print(sum(concat_data['Outlet_Size'].isnull()))
```

```python
print(sum(concat_data["Item_Visibility"]==0))
mean_visibility = concat_data.pivot_table(values="Item_Visibility",index="Item_Identifier")

zero_visibility=(concat_data["Item_Visibility"]==0)
concat_data.loc[zero_visibility,"Item_Visibility"]=concat_data.loc[zero_visibility,"Item_Identifier"].apply(lambda x:mean_visibility.loc[x])
print(sum(concat_data["Item_Visibility"]==0))
```

```python

mean_sale=concat_data.pivot_table(values='Item_Outlet_Sales',columns ='Outlet_Type')
mean_sale
```

```python
concat_data['Item_Visibility_Mean_Ratio'] = concat_data.apply(
    lambda x: x['Item_Visibility']/mean_visibility.loc[x['Item_Identifier']],axis=1)
concat_data['Item_Visibility_Mean_Ratio'].describe()
```

```python
concat_data['Identifier_2']=concat_data.Item_Identifier.str[:2]
concat_data['Item_type_combined']=concat_data.Identifier_2.map({'DR':'Drinks','FD':'Food','NC':'Non-Consumable'})
k=concat_data.groupby('Item_type_combined')
for number,identifier in k:
    print(number," = ",len(identifier))
```

```python
concat_data['Outlet_Years']=2013-concat_data['Outlet_Establishment_Year']
concat_data[['Outlet_Years','Outlet_Establishment_Year']]
```

```python
concat_data.Item_Fat_Content=concat_data.Item_Fat_Content.map({'LF':'Low Fat','low fat':'Low Fat','reg':'Regular','Low Fat':'Low Fat','Regular':'Regular'})
p=concat_data.groupby('Item_Fat_Content')
for fat_type,fat in p:
    print(fat_type," = ",len(fat))
```

```python
miss_val = concat_data['Item_type_combined']=='Non-Consumable'
concat_data.loc[miss_val,'Item_Fat_Content'] = concat_data.loc[miss_val,'Item_type_combined'].apply(lambda x:'Non-Edible')
p=concat_data.groupby('Item_Fat_Content')
for fat_type,fat in p:
    print(fat_type," = ",len(fat))
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
concat_data['Outlet']=encoder.fit_transform(concat_data['Outlet_Identifier'])
columns=['Item_Fat_Content','Item_type_combined','Outlet_Type','Outlet_Size','Outlet_Location_Type','Outlet']

for i in columns:
    concat_data[i]=encoder.fit_transform(concat_data[i])
    
concat_data = pd.get_dummies(concat_data, columns=columns)
concat_data.dtypes
```

```python

concat_data.drop(['Item_Type','Outlet_Establishment_Year'], axis = 1, inplace = True)
concat_data.dtypes
```

```python
a=concat_data.groupby('Source')
for i,j in a:
    if(i=='Test'):
        new_test=j
    else:
        new_train=j
new_test.to_csv("Modified_Test.csv")
new_train.to_csv("Modified_Train.csv")
```

```python
mean_item_sales = new_train.pivot_table(values="Item_Outlet_Sales",index="Item_Identifier")
new_test["Mean_item_Outlet_sale"] = new_test.apply(lambda x: mean_item_sales.loc[x['Item_Identifier']],axis=1)
new_test.to_csv('New_Modified_Test.csv')
```

```python
print(new_test.shape)
print(new_train.shape)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
from sklearn import tree
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor



```

```python

def predict_sale(train,test,alg,predictors,target,new_file_name):
    
    col=['Item_Identifier','Outlet_Identifier']
    col1=['Item_Identifier','Outlet_Identifier','Item_Outlet_Sales','Mean_item_Outlet_sale']
    
    alg.fit(train[predictors],train[target])
    predictions = alg.predict(train[predictors])
    
    cv_score = cross_val_score(alg,train[predictors],train[target],cv=20,scoring="neg_mean_squared_error")
    print("cv=",cv_score)
    cv_score = np.sqrt(np.abs(cv_score))
    
    print("Mean = ",cv_score.mean())
    print("Minimum value  = ",cv_score.min())
    print("Maximum value = ",cv_score.max())
    print("Standard Deviation = ",cv_score.std())

    y_true = train[target].values
    y_pred = predictions
    meanError=np.sqrt(mean_squared_error(y_true, y_pred))
    
    print("mean error =",meanError)
    
    predicted_sale = alg.predict(test[predictors])
    test['Item_Outlet_Sales']=predicted_sale
    new_df_test=test[col1]
    new_df_test.to_csv(new_file_name)
    print()
    return meanError
    
#call function
a=['Item_MRP',  'Item_Visibility', 'Item_Weight',   'Item_Visibility_Mean_Ratio', 'Outlet_Years', 'Item_Fat_Content_0', 'Item_Fat_Content_1', 'Item_Fat_Content_2', 'Item_type_combined_0', 'Item_type_combined_1', 'Item_type_combined_2', 'Outlet_Type_0', 'Outlet_Type_1', 'Outlet_Type_2', 'Outlet_Type_3', 'Outlet_Size_0', 'Outlet_Size_1', 'Outlet_Size_2', 'Outlet_Location_Type_0', 'Outlet_Location_Type_1', 'Outlet_Location_Type_2', 'Outlet_0', 'Outlet_1', 'Outlet_2', 'Outlet_3', 'Outlet_4', 'Outlet_5', 'Outlet_6', 'Outlet_7', 'Outlet_8', 'Outlet_9']
b_target='Item_Outlet_Sales'

mean_error_Linear = predict_sale(new_train,new_test,LinearRegression(normalize=True),a,b_target,"linear.csv")
mean_error_Ridge = predict_sale(new_train,new_test,Ridge(alpha=0.05),a,b_target,"ridge.csv")
mean_error_Decision = predict_sale(new_train,new_test,DecisionTreeRegressor(max_depth=8),a,b_target,"decision.csv")
mean_error_RandomForest = predict_sale(new_train,new_test,RandomForestRegressor(max_depth=15, n_estimators=50),a,b_target,"randomforest.csv")

```

```python
alg=LinearRegression(normalize=True).fit(new_train[predictors],new_train[target])
predictions = alg.predict(new_train[predictors])
plt.figure(figsize=(15,3))
cfs=alg.coef_
h1=pd.Series(cfs,predictors)
h1.plot(kind='bar',title='Model_Coefficients')
```

```python
alg=DecisionTreeRegressor(max_depth=15).fit(new_train[predictors],new_train[target])
predictions = alg.predict(new_train[predictors])
plt.figure(figsize=(15,3))
cfs=alg.feature_importances_
h1=pd.Series(cfs,predictors)
h1.plot(kind='bar',title='Model_Coefficients')
```

```python
alg=RandomForestRegressor(max_depth=15, n_estimators=50).fit(new_train[predictors],new_train[target])
predictions = alg.predict(new_train[predictors])
plt.figure(figsize=(15,3))
cfs=alg.feature_importances_
h1=pd.Series(cfs,predictors)
h1.plot(kind='bar',title='Model_Coefficients')
```

```python
def accuracy(csv_file):
    test=pd.read_csv(csv_file)
    test['Outlet_Score'] = test.apply(lambda x: 1 if x['Item_Outlet_Sales']<0 else 0 if abs(x['Mean_item_Outlet_sale']-x['Item_Outlet_Sales'])<500 else 1, axis=1)
    b=test.groupby('Outlet_Score')['Outlet_Score'].count()
    accuracy=(b[0]/(b[0]+b[1]))*100
    return accuracy

accuracy_linear=accuracy("linear.csv")
accuracy_ridge=accuracy("ridge.csv")
accuracy_decision=accuracy("decision.csv")
accuracy_randomforest=accuracy("randomforest.csv")

print("accuracy_linear = ",accuracy_linear)
print("accuracy_ridge = ",accuracy_ridge)
print("accuracy_decision = ",accuracy_decision)
print("accuracy_randomforest = ",accuracy_randomforest)
```

```python
errors_all=[mean_error_Linear,mean_error_Ridge,mean_error_Decision,mean_error_RandomForest]
accuracy_all=[accuracy_linear,accuracy_ridge,accuracy_decision,accuracy_randomforest]
columns=['Lin','Rr','DT','RF']

fig=plt.figure(figsize=(15,5))
ax1=fig.add_subplot(121)
ax2=fig.add_subplot(122)
plt.rcParams['axes.grid'] = True

ax1.plot(columns,accuracy_all,'o-')
ax2.plot(columns,errors_all,'o-')

ax1.set_title('Accuracy Graph ')
ax1.set_xlabel('Prediction Models')
ax1.set_ylabel('Percentage Accuracy')
ax2.set_title('Mean Squared Error Graph')
ax2.set_xlabel('Prediction Models')
ax2.set_ylabel('Mean Squared Error')
plt.show()

```
